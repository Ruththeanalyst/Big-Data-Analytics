```python
#Load the Dataset with Panda
import pandas as pd

df = pd.read_csv("hygdata_v42.csv.gz")
df.head()
```

```python
#Identify stars that have a "proper" name
df_named = df.dropna(subset=['proper'])
df_named.head()
```

```python
#Sort the rows by distance (descending)and show the first 10
df_sorted = df_named.sort_values(by='dist', ascending=False)
df_sorted.head(10)
```

```python
#The five furthest "proper named" stars
df_sorted.head(5)
```

The data is interesting, but also a bit surprising. Some stars with real names are extremely far away, which makes me wonder if those names are accurate. I also noticed that a few names seem repeated or unusual. It looks like the dataset mixes common star names with scientific catalog names, so everything ends up combined together.

```python
#Further Exploration
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("hygdata_v42.csv.gz")
df.head()
```

```python
#Extension: Exploring More Patterns

clean_df = df[['ci', 'absmag']].dropna()

plt.scatter(clean_df['ci'], clean_df['absmag'])
plt.xlabel("Color Index (ci)")
plt.ylabel("Absolute Magnitude")
plt.title("Color Index vs Absolute Magnitude")
plt.show()
```

This scatter plot shows the relationship between a star’s color index and its absolute magnitude. Each point represents a star. The color index indicates the star’s color (and temperature), while the absolute magnitude shows how bright the star actually is. From the plot, we can see that stars with lower absolute magnitude (brighter stars) tend to have certain color indices, while dimmer stars are spread across other color indices. This pattern reflects different types of stars, such as hot blue stars, cooler red stars, and other stellar groups.
